```python
'''
Car class 작성하기 (Parent)
attribute(멤버변수) :  생성자에 self.name 선언
method              :  def info(self)

(Child1) <- Car 를 상속함
ElectronicCar class
attribute(멤버변수) :  생성자에 self.name 선언
method              :  def info(self) Override
                       이름, 사용연료(electronic) 출력
(Child2) <- Car 를 상속함
GasolineCar class
attribute(멤버변수) :  생성자에 self.name 선언
method              :  def info(self) Override
                       이름, 사용연료(gasoline) 출력
'''
class Car:
    def __init__(self, name):
        self.name = name
    def __str__(self):
        return self.name
    
class ElectronicCar(Car):
    def __str__(self):
        return f"{self.name} 사용연료 - electronic"
    
class GasolineCar(Car):
    def __str__(self):
        return f"{self.name} 사용연료 - gasoline"
    
elec = ElectronicCar("전기차")
gas  = GasolineCar("가솔린차")

print("elec :",elec)
print("gas  :",gas)
        
```

```python
class Car2:
    def __init__(self, name):
        self.name = name
    def display_info(self):
        print(self.name)
    
class ElectronicCar2(Car2):
    def __init__(self, name):
        super().__init__(name)
        # Car2.__init__(self, name)
        # self.name = name
    # display_info() 메소드 overriding
    def display_info(self):
        print(self.name,"사용연료 - electronic")
    
class GasolineCar2(Car2):
    def __init__(self, name):
        super().__init__(name)
        # Car2.__init__(self, name)
        # self.name = name
    # display_info() 메소드 overriding
    def display_info(self):
        print(self.name,"사용연료 - gasoline")
        
elec = ElectronicCar2("전기차")
gas  = GasolineCar2("가솔린차")   

elec.display_info()
gas.display_info()
```

```python
class Car2:
    def __init__(self, name):
        self.name = name
    def display_info(self):
        print(self.name)

class ElectronicCar2(Car2):
    def __init__(self, name, color):
        super().__init__(name)
        # Car2.__init__(self, name)
        self.color = color
    # display_info() 메소드 overriding
    def display_info(self):
        print(self.name,"사용연료 - electronic,","색상 -",self.color)
class GasolineCar2(Car2):
    def __init__(self, name, gearType):
        super().__init__(name)
        # Car2.__init__(self, name)
        self.gearType = gearType
    # display_info() 메소드 overriding
    def display_info(self):
        print(self.name,"사용연료 - gasoline,","기어 -",self.gearType)
        
elec = ElectronicCar2("전기차", "red")
gas  = GasolineCar2("가솔린차", "mannual")   

elec.display_info()
gas.display_info()
```

```python
class Shape:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def display_info(self)    :
        print(f"size : {self.width}, {self.height}")
        
class Rectangle(Shape):
    # display_info() overriding
    def display_info(self):
        print(f"사각형의 넓이 : {self.width * self.height}")
        
class Triangle(Shape):
    # display_info() overriding
    def display_info(self):
        print(f"삼각형의 넓이 : {self.width * self.height * 0.5}")

s1 = Shape(3, 4)        
s1.display_info()
r1 = Rectangle(5, 6)        
r1.display_info()
t1 = Triangle(6, 8)
t1.display_info()
```

```python
from abc import *
class Character(metaclass=ABCMeta):
    @abstractmethod
    def attack(self):
        pass
    @abstractmethod
    def move(self):
        pass
    def say_hello(self):
        print("안녕하세요, 등장인물입니다")
        
class Elf(Character):
    def attack(self):
        print("엘프가 공격함")
    def move(self):
        print("엘프가 이동함")
    def say_hello(self):
        print("안녕하세요, 엘프입니다")
    
class Human(Character):
    def attack(self):
        print("사람이 공격함")
    def move(self):
        print("사람이 이동함")
    def say_hello(self):
        print("안녕하세요, 사람입니다")

elf = Elf()
elf.attack()
elf.move()
elf.say_hello()
print("-" * 40)

human = Human()
human.attack()
human.move()
human.say_hello()
print("-" * 40)   
```

```python
from abc import *

class Person(metaclass=ABCMeta):
    @abstractmethod
    def eat(self):
        pass
    @abstractmethod
    def sleep(self):
        pass

class Student(Person):
    # Overriding
    def eat(self):
        print("학생이 밥을 먹습니다")
    #Overriding
    def sleep(self):
        print("학생이 잠을 잡니다")
    def study(self):
        print("공부를 합니다")
    def go_to_school(self):
        print("학교를 갑니다")
        
class Worker(Person):
    # Overriding
    def eat(self):
        print("밥을 먹습니다")
    #Overriding
    def sleep(self):
        print("잠을 잡니다")
    def work(self):
        print("일을 합니다")
    def go_to_work(self):
        print("출근합니다")
        
s1 = Student()
s1.eat()
s1.sleep()
s1.study()
s1.go_to_school()
print("-" * 40)

w1 = Worker()
w1.eat()
w1.sleep()
w1.work()
w1.go_to_work()
print("-" * 40)
```

```python
from abc import *

class Vehicle(metaclass=ABCMeta):
    def __init__(self, name):
        self.name = name
    @abstractmethod
    def start(self):
        pass
    @abstractmethod
    def stop(self):
        pass
    
class FireFight(Vehicle) :
    def start(self):
        print(f"{self.name}가 불을 끄러 출동합니다")
    def stop(self):
        print(f"{self.name}가 화재 현장에 도착했습니다")
        
class Ambulance(Vehicle):
    def start(self):
        print(f"{self.name}가 환자를 이송하러 출발합니다")
    def stop(self):
        print(f"{self.name}가 병원에 도착했습니다")

class PatrolCar(Vehicle):
    def start(self):
        print(f"{self.name}가 순찰하러 출발합니다")
    def stop(self):
        print(f"{self.name}가 본부에 도착했습니다")
        
class FreightCar(Vehicle):      
    def start(self):
        print(f"{self.name}가 화물을 운반하러 출발합니다")
    def stop(self):
        print(f"{self.name}가 공장에 도착했습니다")
        
        
list_vehicles = []
list_vehicles .append(FireFight("소방차"))
list_vehicles.append(Ambulance("엠뷸런스"))
list_vehicles.append(PatrolCar("순찰차"))
list_vehicles.append(FreightCar("화물차"))

for car in list_vehicles:
    car.start()
    car.stop()   
```

```python
class Calc4:
    def __init__(self):
        self.num1 = 3
        self.num2 = 4
    def add(self):
        print("덧셈 :",self.num1 + self.num2)
    def sub(self):
        print("뺄셈 :",self.num1 - self.num2)
    def multi(self):
        print("곱셈 :",self.num1 * self.num2)
    def div(self):
        print("나눗셈 :",self.num1 / self.num2)
        
class Calc4Power:
    def __init__(self):
        self.num1 = 3
        self.num2 = 4
        self.calc4 = Calc4()
    def power(self):
        print("제곱 :",self.num1 / self.num2)
    
cp1 = Calc4Power()
cp1.calc4.add()
cp1.calc4.sub()
cp1.calc4.multi()
cp1.calc4.div()
cp1.power()
        
```

```python
class Calc1:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    def add(self):
        return self.num1 + self.num2
    def sub(self):
        return self.num1 - self.num2
    
class Calc2:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2        
        # 포함관계 - Calc2 클래스의 멤버변수로 
        #            Calc1 type 의 변수를 선언함
        self.calc1 = Calc1(self.num1, self.num2)
    def multiply(self):
        return self.num1 * self.num2
    def add(self):
        return self.calc1.add()
    def sub(self):
        return self.calc1.sub()

calc2 = Calc2(10, 8)
print(f"calc2.add()      : {calc2.add()}")
print(f"calc2.sub()      : {calc2.sub()}")
print(f"calc2.multiply() : {calc2.multiply()}")
```

```python
def factorial(number):
    if number == 1:
        return 1
    return number * factorial(number - 1)

print("5! :",factorial(5))
```

```python
def power(n1, n2):
    if n2 == 0:
        return 1
    else:
        return (power(n1, n2 - 1) * n1)    
print(power(2, 5))    
```

```python
def fibonacci(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibonacci(n-2) + fibonacci(n-1)
    
print("fibonacci(1) :",fibonacci(1))
print("fibonacci(2) :",fibonacci(2))
print("fibonacci(3) :",fibonacci(3))
print("fibonacci(4) :",fibonacci(4))
print("fibonacci(5) :",fibonacci(5))
print("-" * 40)

number = int(input("피보나치수열의 자릿수를 입력하세요 : "))
print("fibonacci(number) :",fibonacci(number))
```

```python
counter = 0

def fibonacci2(n):
    print(f"fibonacci({n}) 을 구합니다")
    global counter
    counter += 1
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibonacci2(n-2) + fibonacci2(n-1)
    
number = int(input("피보나치수열의 자릿수를 입력하세요 : "))
print("-" * 40)
print("-- fibonacci 함수 내부 --")

print(f"fibonacci2({number}) :",fibonacci2(number))
print("-" * 40)

print(f"fibonacci2({number}) 계산에 활용된 덧셈 횟수는 {counter} 번입니다")
```

```python
counter = 0

def fibonacci3(n):
    # global counter 라고 선언하지 않으면
    # 전역변수 counter 를 사용하지 못함
    counter += 1
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibonacci3(n-2) + fibonacci3(n-1)
    
number = int(input("피보나치수열의 자릿수를 입력하세요 : "))
print("-" * 40)
print("-- fibonacci 함수 내부 --")

print(f"fibonacci3({number}) :",fibonacci3(number))
print("-" * 40)

print(f"fibonacci3({number}) 계산에 활용된 덧셈 횟수는 {counter} 번입니다")
```

```python
dict1 = {
          1:1, 
          2:1
        }

def fibonacci4(n):
    if n in dict1:
        return dict1[n]
    else:
        output   =  fibonacci4(n-1) + fibonacci4(n-2)
        dict1[n] = output
        return output
    
number = int(input("피보나치수열의 자릿수를 입력하세요 : "))
print("-" * 40)
print("-- fibonacci 함수 내부 --")

print(f"fibonacci4({number}) :",fibonacci4(number))
print("-" * 40)

print(f"fibonacci4({number}) 계산에 활용된 덧셈 횟수는 {counter} 번입니다")
```
