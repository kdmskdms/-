```python
'''
클래스 변수             (cf. instance 변수(멤버변수 중 self 가 붙은 것))
  ㄴ self 가 붙지 않음         ㄴ 객체마다 (다른 값을 가지고) 생성됨     
  ㄴ 해당 클래스에 단 하나만 생성됨 - 해당 클래스의 객체들이 공유함
  ㄴ 객체를 생성할 때마다 별도로 생성되지 않음
  ㄴ 메소드 밖에서 self 를 사용하지 않고 작성함
  ㄴ 객체를 생성하기 전에 이미 메모리에 생성되어있음
                                          ㄴ 값을 할당받아서 메모리 상에 존재함
  ㄴ 객체를 생성하기 전부터 메모리에 생성되었기 때문에
     아직 참조변수가 없는 시점임  <--  클래스이름으로 접근함
  ㄴ 객체를 생성하기 전이나 후에 모두 클래스이름으로 접근이 가능함   
'''
```

```python
class Restaurant:
    # 클래스 변수
    list_type = ["한식", "양식", "중식"]
    
    # 생성자
    def __init__(self, name, restaurant_type, address):
        self.name = name
        self.__restaurant_type = restaurant_type
        self.address = address
    
    # getter
    @property
    def restaurant_type(self):
        return self.__restaurant_type
    
    # setter
    @restaurant_type.setter
    def restaurant_type(self, restaurant_type):
        # 클래스 변수는 self. 을 사용하지 않고 클래스이름. 으로 접근해서 사용함
        if restaurant_type in Restaurant.list_type:
            self.__restaurant_type = restaurant_type
            print("식당 종류를 변경함")
        else:
            print("{} 중 하나를 선택하세요".format(", ".join(Restaurant.list_type)))
    def display_info(self):
        return(f"식당이름 : {self.name}\n"
               f"식당종류 : {self.__restaurant_type}\n"
               f"식당주소 : {self.address}")
    
# 식당 객체 생성하기
rest1 = Restaurant("글로벌", "한식", "서울대입구역")
print(rest1.display_info())
print("-" * 50)

rest1.restaurant_type = "양식"
print(rest1.display_info())
print("-" * 50)

rest1.restaurant_type = "분식"
print(rest1.display_info())
print("-" * 50)

# 식당 객체 생성하기
rest2 = Restaurant("아이티", "양식", "사당역")
print(rest2.display_info())
print("-" * 50)

rest2.restaurant_type = "한식"
print(rest2.display_info())
print("-" * 50)
```

```python
class Robot:
    # 클래스 변수
    count_robot = 0
    
    # 생성자 : 객체를 생성할 때 자동으로 호출되는 메소드
    # 클래스 멤버변수 count_robot 값을 1 증가시키고
    # 인스턴스 멤버변수 robot_no 에 증가된 값을 할당함
    #  ㄴ 객체가 생성될 때 마다 인스턴스 변수 robot_no 의 값이 1 증가됨
    def __init__(self):
        Robot.count_robot += 1
        self.robot_no = Robot.count_robot
        
    # 클래스 메소드 : robot 가 몇 대 제작되었는지 출력함 : 총 생산대수
    @classmethod
    def display_count_robot_class(cls):
        print(f"robot 가 {cls.count_robot} 대 제작되었습니다 : in class method")
    
    # 인스턴스 메소드 : 각 robot 별 제작 번호를 출력함
    def display_count_robot_instance(self):
        print(f"robot 제작 번호 : {self.robot_no} 번 : in instance method")
        
# 클래스 메소드 호출하기 : 클래스이름. 으로 접근해서 호출함
Robot.display_count_robot_class()
print("-" * 50)

# Robot 객체 생성하기 - 1
r1 = Robot()
Robot.display_count_robot_class()
r1.display_count_robot_instance()        

# Robot 객체 생성하기 - 2
r2 = Robot()
Robot.display_count_robot_class()
r2.display_count_robot_instance()   

# Robot 객체 생성하기 - 3
r3 = Robot()
Robot.display_count_robot_class()
r3.display_count_robot_instance()   
```

```python
'''
396 page
'''
class Student:
    
    count = 0
    students = []
    
    # 클래스 메소드
    @classmethod
    def print_score(cls):
        print("--------- 학생 목록 ---------")
        print("번호\t이름\t총점\t평균")
        print("-" * 30)
        for student in cls.students:
            print(student)
        print("-" * 30)    
    
    def __init__(self, name, korean, math, english, science):
        self.number = 0
        self.name = name
        self.korean = korean
        self.math = math
        self.english = english
        self.science = science
        
        # 클래스 변수 설정
        Student.count += 1
        self.number = Student.count
        Student.students.append(self)
        
    def get_sum(self):
        return self.korean + self.math + self.english + self.science
    
    def get_average(self):
        return self.get_sum() / 4
    
    # self.number 를 Student.count 로 하면 안 되는 이유
    # 객체를 생성할 때마다 Student.count 가 1 씩 증가되어서
    # Student.print_score() 를 호출하는 시점에서는 
    # Student.count 가 이미 5로 증가되어 있음
    # self.number 으로 하면, 객체 마다 값이 다르므로
    # 학생 번호가 정확하게 출력됨
    def __str__(self):
        return f"{self.number}\t{self.name}\t{self.get_sum()}\t{self.get_average()}"
    
s1 = Student("윤인성",87,98,88,95)
s2 = Student("연하진",92,98,96,98)
s3 = Student("구지연",76,96,94,90)
s4 = Student("나선주",98,92,96,92)
s5 = Student("윤아린",95,98,98,98)

Student.print_score()
```
